Ones and zeros

```python
# 1: Create a new array of 2*2 integers, without initializing entries.
```

```python
arr = np.zeros((2,2), dtype = int)
print(arr)
```

```python
# 2: 
# Let X = np.array([1,2,3], [4,5,6], np.int32). Create a new array with the same shape and type as X.
# Output : array([[1, 2, 3],
#                 [4, 5, 6]])
```

```python
X = np.array([[1,2,3],[4,5,6]], dtype = np.int32)
print(X)
newarr = X.copy()
print(newarr)
```

```python
# 3: Create a 3-D array with ones on the diagonal and zeros elsewhere.
```

```python
x = np.eye(3)
print(x)
```

```python
# 4: Create a new array of 3*2 float numbers, filled with ones.
```

```python
x4 = np.ones((3,2), dtype= np.float64)
print(x4)
```

```python
# 5: Let x = np.arange(4, dtype=np.int64). Create an array of ones with the same shape and type as X.
x = np.arange(4, dtype=np.int64)
newx = np.ones(x.shape, dtype = np.int64)
newx2 = np.ones_like(x)
print(newx)
print(newx2)
```

```python
# 6: Create a new array of 3*2 float numbers, filled with zeros.
```

```python
x = np.zeros((3,2), dtype = np.float64)
print(x)
```

```python
# 7: Let x = np.arange(4, dtype=np.int64). Create an array of zeros with the same shape and type as X.
```

```python
x = np.arange(4, dtype = np.int64)
print(x)
newx = np.zeros(x.shape, int)
print(newx)
newx2 =np.zeros_like(x)
print(newx2)
```
